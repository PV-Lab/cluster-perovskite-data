```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

folder = './Data/XRD'

files = list(os.walk(folder))[0][2]

print('Files to read: ', files)

a = pd.DataFrame()

# Extract XRD x and y component.
for file in files:
  a[file +'_x'] = pd.read_csv(folder + '/' + file, sep=' ', names = ['x', 'y'])['x']
  a[file+'_y'] = pd.read_csv(folder + '/' + file, sep=' ', names = ['x', 'y'])['y']
```

```python
# Investigate the raw data files with and without xlim.

file = files[1]

print(file)

file_df = pd.read_csv(folder + '/' + file, sep=' ', names = ['x', 'y'])
file_df.plot(x='x', y='y')
#plt.xlim(8, 15)
```

```python
# Scale the data
b = a/a.max()
b
```

```python
# Investigate the results of scaling.
cols = ['c1_Cs25FA50MA25.xy_y', 'c2_Cs25FA71MA04.xy_y', 'c3_Cs71FA01MA28.xy_y']

for i in cols:
  print(a[i])
  print(b[i])
```

```python
def plotter(axes, cols, colors, legends, offset=0):
  for i in range(len(cols)):
    col = cols[i]
    axes.plot(a[col+'_x'], b[col+'_y'] + offset, label=legends[i], color=colors[i])
    offset += 1


fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6,6))

cols1 = ['c1_Cs25FA50MA25.xy', 'c2_Cs25FA71MA04.xy', 'c3_Cs71FA01MA28.xy']
legends1 = ['C0: Cs$_{0.25}$FA$_{0.5}$MA$_{0.25}$PbI$_3$', 
           'C1: Cs$_{0.25}$FA$_{0.71}$MA$_{0.04}$PbI$_3$', 
           'C2: Cs$_{0.71}$FA$_{0.01}$MA$_{0.28}$PbI$_3$']
cols2 = ['r1_Cs67FA24MA10.xy', 'r2_Cs37FA62MA01.xy', 'r3_Cs75FA0MA25.xy']
legends2 = ['R0: Cs$_{0.67}$FA$_{0.24}$MA$_{0.1}$PbI$_3$', 
           'R1: Cs$_{0.37}$FA$_{0.62}$MA$_{0.01}$PbI$_3$', 
           'R2: Cs$_{0.75}$MA$_{0.25}$PbI$_3$']

offset = 0

# Let's make the right hand side of the plot lighter.
#colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
#for i in range(len(colors)):
#    colors.append(np.array(mcolors.to_rgb(colors[i])) + 0.2)
#    if i == 1: # Orange gets saturated
#        colors[i+3][0] = 1

# Color cycle for the plots.
colors = ['coral', 'darkorchid', 'darkturquoise', 'magenta']
#lighter_colors =  ['lightcoral', 'orchid', 'turquoise', 'magenta']

plotter(ax1, cols1, colors, legends1)
plotter(ax2, cols2, colors, legends2)

ax1.set_xlim(8, 15)
ax2.set_xlim(8, 15)

ylim = 4
ax1.set_ylim(0, ylim)
ax2.set_ylim(0, ylim)

ax1.tick_params(left=False, labelleft=False)
ax2.tick_params(left=False, labelleft=False)

ax1.set_ylabel('Intensity')

ax1.set_xlabel(r"2$\Theta$")
ax2.set_xlabel(r"2$\Theta$")

height = -0.4
ax1.legend(loc=(0,height), frameon=False)
ax2.legend(loc=(0,height), frameon=False)

ax1.text(10, 3.8, 'Comp. centroids')
ax2.text(10, 3.8, 'RGB centroids')

fontsize = 'small'

for i in range(2):
    if i == 0:
        ax = ax1
    else:
        ax = ax2

    if i==0:
        ax.text(9.9, 3.5, r' $ {\rm \delta}$-Cs (002)', c='b', fontsize=fontsize)
    ax.scatter([9.8], [3.5], marker='*', c='b', s=10)
    ax.scatter([9.8], [1.5], marker='*', c='b', s=10)
    ax.scatter([9.8], [0.5], marker='*', c='b', s=10)
    
    if i==0:
        ax.text(12, 2.9, r' ${\rm \delta}$-Cs (102)', c='b', fontsize=fontsize)
    ax.scatter([13], [2.8], marker='*', c='b', s=10)
    if i == 1:
        ax.scatter([13], [1.3], marker='*', c='b', s=10)
        ax.scatter([13], [0.4], marker='*', c='b', s=10)

    if i==0:
        ax.text(12.3, 3.2, r'perovskite', fontsize=fontsize)
    ax.scatter([14.1], [3.2], marker='d', c='k', s=10)
    ax.plot([14.1, 14.1], [0, ylim], '--k', linewidth=0.8)
    
    if i==0:
        ax.text(12.6, 1.5, r' PbI$_2$', c='darkgoldenrod', fontsize=fontsize)
    ax.scatter([12.6], [1.5], marker='h', c='darkgoldenrod', s=10)
    if i == 1:
        ax.scatter([12.6], [0.5], marker='h', c='darkgoldenrod', s=10)

plt.gcf().set_size_inches(6.5, 4.5)
plt.tight_layout()
fig.savefig('xrd_films_clustered.png', dpi = 600, bbox_inches = 'tight')
fig.savefig('xrd_films_clustered.tiff', dpi = 600, bbox_inches = 'tight')
fig.savefig('xrd_films_clustered.svg', bbox_inches = 'tight')
```
